```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
from tensorflow import keras
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
xi = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
yi = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xi, yi, epochs=500)
```

```python
print(model.predict([10.0]))
```

```python
print(model.predict([1.0]))
```

```python
print(model.predict([-1.0]))
```

```python
house_model = keras.Sequential(keras.layers.Dense(units=3, input_shape=[1]))
```

```python
house_model.compile(optimizer='sgd', loss="mean_squared_error")
```

```python
bedrooms = np.array([1.0, 18.0, 5.0, 3.0, 10.0, 15.0, 8.0, 13.0], dtype=float)
prices = np.array([1.0, 9.5, 3.0, 2.0, 5.5, 8.0, 4.5, 7.0], dtype=float)
```

```python
house_model.fit(bedrooms, prices, epochs=1000)
```

```python
print(np.mean(house_model.predict([7.0])[0]*100000))
```

```python
fminst = tf.keras.datasets.fashion_mnist
```

```python
(training_images, training_labels), (test_images, test_labels) = fminst.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(training_images[159])
```

```python
training_images = training_images / 255
test_images = test_images / 255
```

```python
fashion_model = tf.keras.Sequential([tf.keras.layers.Flatten(),
                                 tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
fashion_model.compile(optimizer=tf.optimizers.Adam(), loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])
```

```python
fashion_model.fit(training_images, training_labels, epochs=5)
```

```python
fashion_model.evaluate(test_images, test_labels)
```

```python
classifications = fashion_model.predict(test_images)
```
